```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("Salary Data.csv")
```

```python
df.head()
```

```python
X = df["YearsExperience"]
y = df["Salary"]
```

# Creating The Model

```python
def f(x_i,w,b):
    return w*x_i+b
```

```python
def cost_function(w,b,X,y):
    s = 0
    m = len(X)
    for i in range(m):
       s += (f(X[i],w,b) - y[i])**2
    s = s/(2*m)
    return s
       
```

```python
def gradient(w,b,X,y):
    d_w,d_b = 0,0
    m = len(X)
    for i in range(m):
        erreur = f(X[i],w,b)-y[i]
        d_w += erreur*X[i]
        d_b += erreur
    return (d_w/m,d_b/m)
    
```

```python
def gradient_descent(w,b,X,y,learning_rate,iterations):
    for i in range(iterations):
        d_w,d_b = gradient(w,b,X,y)
        w = w- learning_rate*d_w
        b = b- learning_rate*d_b
    print(f"MSE : {cost_function(w,b,X,y)}")
    return w,b
```

```python
w,b = gradient_descent(0, 0, X, y, 0.01, 1000)
```

# Visualisation

```python
y_pred = [f(X[i], w, b) for i in range(len(X))]
plt.scatter(X, y, color='blue', label='Real data')
plt.plot(X, y_pred, color='red', label='expected')
plt.legend()
plt.show()
```
